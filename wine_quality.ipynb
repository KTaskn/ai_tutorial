# 演習

## データ内容
* _class: red/white
* fixed acidity: 酒石酸濃度
* volatile acidity: 酢酸濃度
* citric acid: クエン酸濃度
* residual sugar: 残糖濃度
* chlorides: 塩化ナトリウム濃度
* free sulfur dioxide: 遊離SO2（二酸化硫黄）濃度
* total sulfur dioxide:総SO2（二酸化硫黄）濃度
* density: 密度
* pH: 水素イオン濃度
* sulphates: 硫化カリウム濃度
* alcohol: アルコール度数
* quality: **評価**

qualityの値を予測しよう

```python
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
FILE_LEARN = 'dataset/winequality_learn.csv'
FILE_PREDICT = 'dataset/winequality_predict.csv'

df_learn = pd.read_csv(FILE_LEARN)
display(df_learn.head(5))
df_predict = pd.read_csv(FILE_PREDICT)
display(df_predict.head(5))
```

# 可視化
わかりやすい可視化に箱ひげ図（ボックスプロット）がある

```python
for col in df_learn.drop(['quality', '_class', 'id'], axis=1).columns:
    for c, df_cls in df_learn.groupby('_class'):
            l_plt = []
            for idx, df_grp in df_cls.groupby('quality'):
                l_plt.append(df_grp[col])
            print(c, col)
            plt.figure(figsize=(7, 7))
            plt.boxplot(l_plt)
            plt.show()
```

# 回帰係数を利用したテクニック
変数が非常に多い場合には回帰係数を利用してあたりをつけることができる

```python
lr = LinearRegression()

COL_CLASS = '_class'
COL_TARGET = 'quality'
COL_FEATURE = df_learn.drop(['id', COL_TARGET, COL_CLASS], axis=1).columns

# Z-Scoreを利用して、すべての変数を同じ幅（-1〜1）ぐらいにする
def zscore(X):
    return (X - X.mean()) / X.std()

X = df_learn[COL_FEATURE].apply(zscore)
y = df_learn[COL_TARGET]
lr.fit(X, y)

# 係数が影響度の目安になる
display(
    pd.DataFrame({
        'name': COL_FEATURE,
        'coef': lr.coef_,
        'importance': np.abs(lr.coef_)
    }).sort_values('importance')
)

df_learn['predict'] = lr.predict(X)

l_plot = []
for idx, df_grp in df_learn.groupby('quality'):
    l_plot.append(df_grp['predict'])

plt.figure(figsize=(6, 6))
plt.boxplot(l_plot)
plt.show()

r2_score(df_learn['quality'], df_learn['predict'])
```
